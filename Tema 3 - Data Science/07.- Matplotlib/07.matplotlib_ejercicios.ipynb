### Ejercicio 4:
- Obtener el histograma de la columna _**Ancho Petalo**_ dataset de la **iris**.

<div align="center">
   
</div>

```python
df = pd.read_csv('../Data/iris.csv')
df.head
```

```python
# Histograma con matplotlib
plt.hist(df['Ancho Petalo'], bins=10, color='blue')
plt.show()
```

```python
# Histograma con pandas (Tambien usa matplotlib)
df['Ancho Petalo'].hist(bins=10, color='blue')
plt.grid(visible=False) # No muestra la cuadricula
plt.show()
```

### Ejercicio 5:
- Obtener un histograma con las columnas _**Largo Petalo**_ y _**Ancho Petalo**_ del dataset de **iris**.

<div align="center">
   
</div>

```python
# Por separado
plt.figure(figsize=(3,3)) # Tamaño del diagrama
plt.hist(df['Ancho Petalo'], alpha = 0.5)
plt.show()
plt.figure(figsize=(3,3))
plt.hist(df['Largo Petalo'], alpha = 0.5)
plt.show()
```

```python
# Los 2 juntos en horizontal, para quitarlo no se pone lo de horizontal
plt.hist(df['Largo Petalo'], alpha = 0.5, orientation='horizontal')
plt.hist(df['Ancho Petalo'], alpha = 0.5, rientation='horizontal')
plt.show()
```

```python

plt.subplot(1, 2, 1)
plt.hist(df['Largo Petalo'], alpha = 0.5)

plt.subplot(1, 2, 2)
plt.hist(df['Ancho Petalo'], alpha = 0.5)

plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.hist(df['Largo Petalo'], alpha = 0.5)

plt.subplot(2, 1, 2)
plt.hist(df['Ancho Petalo'], alpha = 0.5)

plt.show()
```

### Ejercicio 9:
- Muestra en una gráfica de sectores el número de los que no sobrevivieron por clase (_**Pclass**_).

<div align="center">
   
</div>

```python
df[df['Survived'] == 0].value_counts('Pclass', sort=False) # Filtro de los que muerieron
```

```python
values = df[df['Survived'] == 0].value_counts('Pclass', sort=False) # Filtro de los que muerieron
labels= sorted([f"clase {i}" for i in df['Pclass'].unique()])
plt.pie(x=values, 
        labels=labels, #etiquetas
        shadow=True, #sombreado
        colors=['cyan', 'red', 'pink'], #colores
        explode= (0.05, 0.05, 0.05), #separacion de los trozos
        autopct="%1.2f%%"); #numeros de porcenjate
```
