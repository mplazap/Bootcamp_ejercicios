## **2ï¸âƒ£ CreaciÃ³n de Arrays en NumPy**

Los arrays en NumPy son mÃ¡s rÃ¡pidos y eficientes que las listas de Python.

### **ğŸ”¹ Convertir una Lista en un `ndarray`**
```python
import numpy as np

lista = [1, 2, 3, 4, 5]
arr = np.array(lista)
print(arr)  # âœ… [1 2 3 4 5]
print(type(arr))  # âœ… <class 'numpy.ndarray'>
```

### **ğŸ”¹ Arrays Multidimensionales**
```python
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print(matriz)
```
ğŸ“Œ **Salida:**
```
[[1 2 3]
 [4 5 6]]
```

### **ğŸ”¹ Arrays de Zeros y Ones**
```python
zeros = np.zeros((3, 3))  # Matriz 3x3 de ceros
ones = np.ones((2, 4))  # Matriz 2x4 de unos
print(zeros, ones)
```

### **ğŸ”¹ `arange()` y `linspace()`**
```python
arr1 = np.arange(0, 10, 2)  # [0 2 4 6 8]
arr2 = np.linspace(0, 5, 10)  # 10 valores equidistantes entre 0 y 5
```

### **ğŸ”¹ Matriz Identidad (`eye()`)**
```python
identidad = np.eye(4)  # Matriz identidad 4x4
```

## **3ï¸âƒ£ Propiedades de un `ndarray`**
```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr.shape)  # âœ… (2, 3) â†’ Dimensiones
print(arr.ndim)  # âœ… 2 â†’ NÃºmero de dimensiones
print(arr.size)  # âœ… 6 â†’ NÃºmero total de elementos
print(arr.dtype)  # âœ… int64 (depende del sistema)
```

---

## **4ï¸âƒ£ Indexado y Slicing en NumPy**
```python
arr = np.array([10, 20, 30, 40, 50])

print(arr[1])  # âœ… 20 (Acceder al segundo elemento)
print(arr[1:4])  # âœ… [20 30 40] (Slicing)
print(arr[::-1])  # âœ… [50 40 30 20 10] (Invertir array)
```

ğŸ“Œ **Indexado en Matrices**:
```python
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print(matriz[0, 1])  # âœ… 2 (Fila 0, columna 1)
print(matriz[:, 1])  # âœ… [2 5] (Segunda columna)
```

## **5ï¸âƒ£ Operaciones MatemÃ¡ticas con Arrays**

NumPy permite **realizar operaciones matemÃ¡ticas vectorizadas** sin necesidad de bucles.

```python
arr = np.array([1, 2, 3, 4, 5])

print(arr + 10)  # âœ… [11 12 13 14 15]
print(arr * 2)  # âœ… [2 4 6 8 10]
print(arr ** 2)  # âœ… [1 4 9 16 25]
```

ğŸ“Œ **Operaciones entre Arrays**:
```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(arr1 + arr2)  # âœ… [5 7 9]
print(arr1 * arr2)  # âœ… [4 10 18]
```

---

## **6ï¸âƒ£ Funciones MatemÃ¡ticas en NumPy**
```python
arr = np.array([0, np.pi / 2, np.pi])

print(np.sin(arr))  # âœ… Seno de cada elemento
print(np.log(arr + 1))  # âœ… Logaritmo natural (evitar log(0))
print(np.exp(arr))  # âœ… Exponencial e^x
```

ğŸ“Œ **Funciones EstadÃ­sticas**:
```python
print(np.mean(arr))  # âœ… Media
print(np.median(arr))  # âœ… Mediana
print(np.std(arr))  # âœ… DesviaciÃ³n estÃ¡ndar
```

---

## **7ï¸âƒ£ Ãlgebra Lineal en NumPy**
```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(np.dot(A, B))  # âœ… Producto Matricial
print(np.linalg.det(A))  # âœ… Determinante de A
print(np.linalg.inv(A))  # âœ… Inversa de A
```

# **9ï¸âƒ£ Encoding en Pandas**

Los modelos de machine learning requieren datos **numÃ©ricos**, por lo que es necesario **convertir variables categÃ³ricas** en nÃºmeros. Este proceso se llama **Encoding**.

## **1ï¸âƒ£0ï¸âƒ£ One-Hot Encoding (`get_dummies()`)**
âœ” **Ãštil para variables categÃ³ricas sin orden**  
âœ” **Crea una columna para cada categorÃ­a con 0 o 1**  

```python
df_encoded = pd.get_dummies(df, columns=["Ciudad"], drop_first=True)
```

ğŸ“Œ **Ejemplo**:
```
      Ciudad  Ciudad_Barcelona  Ciudad_Madrid  Ciudad_Sevilla
0    Madrid                0              1               0
1  Barcelona               1              0               0
2   Sevilla                0              0               1
```

## **1ï¸âƒ£1ï¸âƒ£ Label Encoding (`LabelEncoder()`)**
âœ” **Asigna un nÃºmero Ãºnico a cada categorÃ­a**  
âœ” **Ãštil para variables ordinales (con orden lÃ³gico)**  

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df["Genero_encoded"] = encoder.fit_transform(df["Genero"])
```

ğŸ“Œ **Ejemplo**:
```
  Genero  Genero_encoded
0      M               1
1      F               0
2      F               0
3      M               1
```

## **1ï¸âƒ£2ï¸âƒ£ Ordinal Encoding (`OrdinalEncoder()`)**
âœ” **Ãštil para variables con jerarquÃ­a (pequeÃ±o < mediano < grande)**  

```python
from sklearn.preprocessing import OrdinalEncoder

orden = OrdinalEncoder(categories=[["PequeÃ±o", "Mediano", "Grande"]])
df["Talla_encoded"] = orden.fit_transform(df[["Talla"]])
```

ğŸ“Œ **Ejemplo**:
```
         Talla  Talla_encoded
0    PequeÃ±o            0.0
1    Grande            2.0
2   Mediano            1.0
```

## **1ï¸âƒ£3ï¸âƒ£ Target Encoding (Mean Encoding)**
âœ” **Usa la media de la variable objetivo para codificar categorÃ­as**  
âœ” **Ãštil en modelos de Ã¡rboles como XGBoost y Random Forest**  

```python
df["Ciudad_encoded"] = df.groupby("Ciudad")["Salario"].transform("mean")
```

ğŸ“Œ **Ejemplo**:
```
      Ciudad  Ciudad_encoded
0    Madrid           3100.0
1  Barcelona          2650.0
2   Sevilla           2700.0
```

## **1ï¸âƒ£4ï¸âƒ£ Hash Encoding (Feature Hashing)**
âœ” **Reduce dimensionalidad cuando hay muchas categorÃ­as**  

```python
from sklearn.feature_extraction import FeatureHasher

fh = FeatureHasher(n_features=3, input_type="string")
hashed_features = fh.fit_transform(df["Ciudad"])
print(hashed_features.toarray())
```

ğŸ“Œ **Ejemplo**:
```
[[ 0.  0.  1.]
 [ 1.  0.  0.]
 [ 0.  1.  0.]]
```
