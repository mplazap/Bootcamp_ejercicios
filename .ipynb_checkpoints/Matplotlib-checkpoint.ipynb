# **3ï¸âƒ£ Tipos de GrÃ¡ficos en Matplotlib**

### **ğŸ”¹ GrÃ¡fico de LÃ­nea (`plot()`)**
âœ” **Uso:** Representa **tendencias en el tiempo** o relaciones entre dos variables numÃ©ricas.  
âœ” **Ejemplo:** EvoluciÃ³n de las ventas mensuales.

```python
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
ventas = [5000, 5200, 6000, 6500, 7000, 7200]

plt.plot(meses, ventas, marker="o", linestyle="-", color="b", label="Ventas")
plt.title("EvoluciÃ³n de Ventas")
plt.xlabel("Meses")
plt.ylabel("Ventas ($)")
plt.legend()
plt.show()
```
ğŸ“Œ **Ejemplo**:
![GrÃ¡fico de LÃ­nea](imagenes/line_plot.png)

---

### **ğŸ”¹ GrÃ¡fico de DispersiÃ³n (`scatter()`)**
âœ” **Uso:** Representa **correlaciones entre dos variables** sin una conexiÃ³n entre los puntos.  
âœ” **Ejemplo:** RelaciÃ³n entre la publicidad y las ventas.

```python
publicidad = [1000, 1500, 2000, 2500, 3000, 3500]
ventas = [7000, 7500, 8500, 9000, 9500, 10000]

plt.scatter(publicidad, ventas, color="blue", label="Publicidad vs Ventas")
plt.title("Impacto de la Publicidad en Ventas")
plt.xlabel("Gasto en Publicidad ($)")
plt.ylabel("Ventas ($)")
plt.legend()
plt.show()
```
ğŸ“Œ **Ejemplo**:
![GrÃ¡fico de DispersiÃ³n](imagenes/scatter_plot.png)

---

### **ğŸ”¹ GrÃ¡fico de Barras (`bar()`)**
âœ” **Uso:** Comparar valores entre diferentes categorÃ­as.  
âœ” **Ejemplo:** Ventas por sucursal.

```python
sucursales = ["Madrid", "Barcelona", "Sevilla", "Valencia"]
ventas = [15000, 18000, 12000, 13000]

plt.bar(sucursales, ventas, color="green")
plt.title("Ventas por Sucursal")
plt.xlabel("Sucursal")
plt.ylabel("Ventas ($)")
plt.show()
```
ğŸ“Œ **Ejemplo**:
![GrÃ¡fico de Barras](imagenes/bar_plot.png)

---

### **ğŸ”¹ Histograma (`hist()`)**
âœ” **Uso:** Mostrar la distribuciÃ³n de un conjunto de datos.  
âœ” **Ejemplo:** DistribuciÃ³n de edades en una empresa.

```python
import numpy as np

edades = np.random.normal(35, 10, 1000)

plt.hist(edades, bins=20, color="purple", edgecolor="black")
plt.title("DistribuciÃ³n de Edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```
ğŸ“Œ **Ejemplo**:
![Histograma](imagenes/histogram.png)

---

### **ğŸ”¹ GrÃ¡fico de Torta (`pie()`)**
âœ” **Uso:** Representar proporciones dentro de un conjunto de categorÃ­as.  
âœ” **Ejemplo:** DistribuciÃ³n de clientes por canal de venta.

```python
canales = ["Online", "Tienda FÃ­sica", "Distribuidores", "Mayoristas"]
porcentajes = [40, 35, 15, 10]

plt.pie(porcentajes, labels=canales, autopct="%1.1f%%", colors=["red", "blue", "green", "orange"])
plt.title("DistribuciÃ³n de Clientes por Canal de Venta")
plt.show()
```
ğŸ“Œ **Ejemplo**:
![GrÃ¡fico de Torta](imagenes/pie_chart.png)

# **5ï¸âƒ£ GrÃ¡ficos 3D con Matplotlib**

âœ” **Uso:** Representar datos con tres dimensiones.  
âœ” **Ejemplo:** RepresentaciÃ³n del trÃ¡fico web en diferentes horarios.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

horas = [1, 2, 3, 4, 5]
usuarios = [100, 200, 150, 250, 300]
compras = [10, 15, 8, 18, 22]

ax.scatter(horas, usuarios, compras)
ax.set_title("TrÃ¡fico Web y Compras")
ax.set_xlabel("Horas")
ax.set_ylabel("Usuarios Conectados")
ax.set_zlabel("Compras Realizadas")

plt.show()
```
ğŸ“Œ **Ejemplo**:
![GrÃ¡fico 3D](imagenes/3d_plot.png)

# **6ï¸âƒ£ IntegraciÃ³n con Pandas**

âœ” **Uso:** Crear grÃ¡ficos directamente desde un `DataFrame`.  
âœ” **Ejemplo:** ComparaciÃ³n de ingresos anuales.

```python
import pandas as pd

df = pd.DataFrame({"AÃ±o": [2020, 2021, 2022], "Ingresos": [50000, 60000, 70000]})
df.plot(x="AÃ±o", y="Ingresos", kind="bar", legend=False)
plt.title("EvoluciÃ³n de Ingresos")
plt.xlabel("AÃ±o")
plt.ylabel("Ingresos ($)")
plt.show()
```
ğŸ“Œ **Ejemplo**:
![GrÃ¡fico de Pandas](imagenes/pandas_bar_chart.png)
