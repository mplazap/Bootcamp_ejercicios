## **2ï¸âƒ£ CreaciÃ³n de Arrays en NumPy**

Los arrays en NumPy son mÃ¡s rÃ¡pidos y eficientes que las listas de Python.

### **ğŸ”¹ Convertir una Lista en un `ndarray`**
```python
import numpy as np

lista = [1, 2, 3, 4, 5]
arr = np.array(lista)
print(arr)  # âœ… [1 2 3 4 5]
print(type(arr))  # âœ… <class 'numpy.ndarray'>
```

### **ğŸ”¹ Arrays Multidimensionales**
```python
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print(matriz)
```
ğŸ“Œ **Salida:**
```
[[1 2 3]
 [4 5 6]]
```

### **ğŸ”¹ Arrays de Zeros y Ones**
```python
zeros = np.zeros((3, 3))  # Matriz 3x3 de ceros
ones = np.ones((2, 4))  # Matriz 2x4 de unos
print(zeros, ones)
```

### **ğŸ”¹ `arange()` y `linspace()`**
```python
arr1 = np.arange(0, 10, 2)  # [0 2 4 6 8]
arr2 = np.linspace(0, 5, 10)  # 10 valores equidistantes entre 0 y 5
```

### **ğŸ”¹ Matriz Identidad (`eye()`)**
```python
identidad = np.eye(4)  # Matriz identidad 4x4
```

## **3ï¸âƒ£ Propiedades de un `ndarray`**
```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr.shape)  # âœ… (2, 3) â†’ Dimensiones
print(arr.ndim)  # âœ… 2 â†’ NÃºmero de dimensiones
print(arr.size)  # âœ… 6 â†’ NÃºmero total de elementos
print(arr.dtype)  # âœ… int64 (depende del sistema)
```

---

## **4ï¸âƒ£ Indexado y Slicing en NumPy**
```python
arr = np.array([10, 20, 30, 40, 50])

print(arr[1])  # âœ… 20 (Acceder al segundo elemento)
print(arr[1:4])  # âœ… [20 30 40] (Slicing)
print(arr[::-1])  # âœ… [50 40 30 20 10] (Invertir array)
```

ğŸ“Œ **Indexado en Matrices**:
```python
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print(matriz[0, 1])  # âœ… 2 (Fila 0, columna 1)
print(matriz[:, 1])  # âœ… [2 5] (Segunda columna)
```

---

## **5ï¸âƒ£ Operaciones MatemÃ¡ticas con Arrays**

NumPy permite **realizar operaciones matemÃ¡ticas vectorizadas** sin necesidad de bucles.

```python
arr = np.array([1, 2, 3, 4, 5])

print(arr + 10)  # âœ… [11 12 13 14 15]
print(arr * 2)  # âœ… [2 4 6 8 10]
print(arr ** 2)  # âœ… [1 4 9 16 25]
```

ğŸ“Œ **Operaciones entre Arrays**:
```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(arr1 + arr2)  # âœ… [5 7 9]
print(arr1 * arr2)  # âœ… [4 10 18]
```

---

## **6ï¸âƒ£ Funciones MatemÃ¡ticas en NumPy**
```python
arr = np.array([0, np.pi / 2, np.pi])

print(np.sin(arr))  # âœ… Seno de cada elemento
print(np.log(arr + 1))  # âœ… Logaritmo natural (evitar log(0))
print(np.exp(arr))  # âœ… Exponencial e^x
```

ğŸ“Œ **Funciones EstadÃ­sticas**:
```python
print(np.mean(arr))  # âœ… Media
print(np.median(arr))  # âœ… Mediana
print(np.std(arr))  # âœ… DesviaciÃ³n estÃ¡ndar
```

---

## **7ï¸âƒ£ Ãlgebra Lineal en NumPy**
```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(np.dot(A, B))  # âœ… Producto Matricial
print(np.linalg.det(A))  # âœ… Determinante de A
print(np.linalg.inv(A))  # âœ… Inversa de A
```

## **8ï¸âƒ£ GeneraciÃ³n de NÃºmeros Aleatorios (`numpy.random`)**

### **ğŸ”¹ Generar un NÃºmero Aleatorio**
```python
print(np.random.rand())  # âœ… NÃºmero aleatorio entre 0 y 1
```

### **ğŸ”¹ Generar un Array Aleatorio**
```python
print(np.random.rand(3, 3))  # âœ… Matriz 3x3 de nÃºmeros entre 0 y 1
```

### **ğŸ”¹ NÃºmeros Enteros Aleatorios**
```python
print(np.random.randint(1, 100, (3, 3)))  # âœ… Matriz 3x3 de enteros entre 1 y 100
```

---

## **9ï¸âƒ£ ManipulaciÃ³n de Arrays**

### **ğŸ”¹ Transponer una Matriz**
```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.T)  # âœ… Transpuesta
```

### **ğŸ”¹ Apilar Arrays**
```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.vstack((a, b)))  # âœ… Apilar verticalmente
print(np.hstack((a, b)))  # âœ… Apilar horizontalmente
```
